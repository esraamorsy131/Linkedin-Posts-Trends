# Identify The Trends In Linkedin Posting

```python
# Number of posts for each profile

Profile_posts_count = df[['Full name','Uploaded on']].groupby(by='Full name').count().reset_index()

Profile_posts_count.sort_values(by='Uploaded on', ascending = False , inplace = True)

Profile_posts_count.rename(columns = {'Uploaded on':'Number Of Posts'} , inplace = True)



Profile_posts_count

```

```python
# Highlight the top 2 Profiles Have Posts

def highlight_top_2(row):
    return ['background-color: yellow' if row.name in Profile_posts_count.head(2).index else '' for _ in row]

# Apply the styling function
styled_df = Profile_posts_count.style.apply(highlight_top_2, axis=1)

# Display the styled DataFrame
styled_df
```

```python


# Extract the year and month directly if column is date type
df['Year'] = df['Date GMT'].apply(lambda x: x.year)
df['Month'] = df['Date GMT'].apply(lambda x: x.strftime('%B'))


# Extract the hour from time objects
df['Hour Of Day'] = df['Time GMT'].apply(lambda x: x.hour)


# Extract the Week Number
df['Week Number'] = df['Uploaded on'].dt.week

df[['Time GMT' , 'Hour Of Day']]
```

# Frequency of Posts
How often a user posts, e.g., daily, weekly, or monthly. 
Some users may post consistently, while others may have sporadic posting behavior.

```python
# Number of posts for each profile (Year To Day)

Profile_posts_count_YTD = df[['Full name','Year','Month','Week Number','Day of week','Hour Of Day','Uploaded on','Date GMT']].groupby(by=['Full name','Year','Month','Week Number','Date GMT','Day of week','Hour Of Day']).count().reset_index()

Profile_posts_count_YTD.sort_values(by = ['Full name','Month'] , inplace = True)

Profile_posts_count_YTD.rename(columns = {'Uploaded on' : 'Number Of Posts'}, inplace = True)

Profile_posts_count_YTD
```

# Best Times to Post:
Identifying the most active times of day and days of the week when users post. 
This can help in understanding when engagement is highest.

```python
# most active  days of the week when users post

def get_most_active_day_for_profile(data):
    # Group by 'Full name' and 'Week GMT' to calculate total posts per day for each profile
    daily_posts = data.groupby(['Full name', 'Day of week'])['Number Of Posts'].sum().reset_index()
    
    # Find the day with the maximum number of posts for each profile
    max_posts_per_profile = daily_posts.loc[daily_posts.groupby('Full name')['Number Of Posts'].idxmax()]

    # Extract relevant information
    most_active_days = max_posts_per_profile[['Full name','Day of week', 'Number Of Posts']]
    most_active_days.rename(columns={'Day of week': 'Most Active Day', 'Number Of Posts': 'Max Number Of Posts'}, inplace=True)
    
    return most_active_days

# Apply the function to get the most active day for each profile
most_active_days = get_most_active_day_for_profile(Profile_posts_count_YTD)
most_active_days_df = most_active_days[most_active_days['Max Number Of Posts'] > 1]

# Print the result
most_active_days_df
```

```python
# most active times of day when users post

def get_most_active_Hour_for_profile(data):
    # Group by 'Full name' and 'Week GMT' to calculate total posts per day for each profile
    daily_posts = data.groupby(['Full name', 'Hour Of Day'])['Number Of Posts'].sum().reset_index()
    
    # Find the day with the maximum number of posts for each profile
    max_posts_per_profile = daily_posts.loc[daily_posts.groupby('Full name')['Number Of Posts'].idxmax()]

    # Extract relevant information
    most_active_Hour = max_posts_per_profile[['Full name','Hour Of Day', 'Number Of Posts']]
    most_active_Hour.rename(columns={'Hour Of Day': 'Most Active Hour', 'Number Of Posts': 'Max Number Of Posts'}, inplace=True)
    
    return most_active_Hour

# Apply the function to get the most active day for each profile
most_active_Hours = get_most_active_Hour_for_profile(Profile_posts_count_YTD)
most_active_Hours_df = most_active_Hours[most_active_Hours['Max Number Of Posts'] > 1]

# Print the result
most_active_Hours_df 
```

# Data Visualization 

```python
# Group by Day of Week and Hour of Day, and count posts
heatmap_data = df.groupby(['Day of week', 'Hour Of Day']).size().unstack(fill_value=0)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', cbar_kws={'label': 'Number of Posts'})

# Add labels and title
plt.title('Heatmap of Post Frequency by Day of Week and Hour of Day')
plt.xlabel('Hour Of Day')
plt.ylabel('Day of week')



# Show the plot

plt.tight_layout()

plt.show()


```

```python
# Group by hour of the day and count posts
posts_by_hour = df.groupby('Hour Of Day').size()

# Create the line plot
plt.figure(figsize=(12, 8))
sns.lineplot(x=posts_by_hour.index, y=posts_by_hour.values, marker='o', color='b')

# Add labels and title
plt.title('Posts by Hour of Day')
plt.xlabel('Hour Of Day')
plt.ylabel('Number of Posts')

# Show the plot
plt.grid(True)
plt.xticks(range(0, 24, 1))  # Hourly ticks

plt.show()



```

```python
# Group by day of the week and count posts
posts_by_day = df.groupby('Day of week').size()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=posts_by_day.index, y=posts_by_day.values, palette='Set2')

# Add labels and title
plt.title('Posts by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Posts')

# Show the plot
plt.show()
```
